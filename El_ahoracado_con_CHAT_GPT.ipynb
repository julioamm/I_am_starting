```python

import random

def seleccion_palabra_secreta():
    estados_mexicanos = [
        "Aguascalientes", "Baja California", "Baja California Sur", "Campeche", "Chiapas",
        "Chihuahua", "Coahuila", "Colima", "Durango", "Estado de Mexico", "Guanajuato",
        "Guerrero", "Hidalgo", "Jalisco", "Michoacan", "Morelos", "Nayarit", "Nuevo Leon",
        "Oaxaca", "Puebla", "Queretaro", "Quintana Roo", "San Luis Potosi", "Sinaloa",
        "Sonora", "Tabasco", "Tamaulipas", "Tlaxcala", "Veracruz", "Yucatan", "Zacatecas"
    ]
    palabra_secreta = random.choice(estados_mexicanos)
    palabra_secreta = palabra_secreta.lower()  # Convierte a minúsculas y quita acentos
    return palabra_secreta

def imprimir_info(palabra_oculta, letras_seleccionadas, intentos_restantes):
    print(f"Palabra oculta: {palabra_oculta}")
    print(f"Letras seleccionadas: {', '.join(letras_seleccionadas)}")
    print(f"Intentos restantes: {intentos_restantes}")

def seleccion_de_letra(letras_seleccionadas):
    while True:
        letra_seleccionada = input("Selecciona una letra del alfabeto para saber si se encuentra en la 'Palabra secreta': ").lower()
        if letra_seleccionada.isalpha() and len(letra_seleccionada) == 1 and 'a' <= letra_seleccionada <= 'z':
            if letra_seleccionada not in letras_seleccionadas:
                return letra_seleccionada
            else:
                print("Ya seleccionaste esa letra. Intenta con otra.")
        else:
            print("Por favor, ingresa una letra válida del alfabeto.")

def juego():
    palabra_secreta = seleccion_palabra_secreta()
    letras_acertadas = []
    letras_seleccionadas = []
    intentos_maximos = 6
    intentos_realizados = 0

    while intentos_realizados < intentos_maximos:
        letra_seleccionada = seleccion_de_letra(letras_seleccionadas)
        letras_seleccionadas.append(letra_seleccionada)

        coincidencias = [i for i, letra in enumerate(palabra_secreta) if letra == letra_seleccionada]

        if coincidencias:
            print(f"¡Bien hecho! La letra '{letra_seleccionada}' está en la palabra secreta en las posiciones {coincidencias}.")
            letras_acertadas.extend(coincidencias)
        else:
            print(f"La letra '{letra_seleccionada}' no está en la palabra secreta. Intenta de nuevo.")
            intentos_realizados += 1

        palabra_oculta = ocultar_palabra(palabra_secreta, letras_acertadas)
        imprimir_info(palabra_oculta, letras_seleccionadas, intentos_maximos - intentos_realizados)

        if set(letras_acertadas) == set(range(len(palabra_secreta))):
            print("¡Felicidades! Has adivinado la palabra secreta.")
            break

        if intentos_realizados == intentos_maximos:
            print("Lo siento, has agotado todos tus intentos. La palabra secreta era:", palabra_secreta)
            break

# Función para ocultar la palabra
def ocultar_palabra(palabra, letras_acertadas):
    return ''.join([letra if i in letras_acertadas else '*' for i, letra in enumerate(palabra)])

# Ejecutar el juego
juego()
```

Explicación detallada del código:

### Función `seleccion_palabra_secreta()`:
- Esta función elige una palabra secreta al azar del conjunto de estados mexicanos.
- Convierte la palabra a minúsculas para hacer el juego insensible a mayúsculas y minúsculas.
- Retorna la palabra secreta.

### Función `ocultar_palabra(palabra, letras_acertadas)`:
- Esta función recibe la palabra secreta y la lista de letras acertadas.
- Devuelve una cadena donde las letras acertadas se muestran y el resto se ocultan con asteriscos.

### Función `imprimir_info(palabra_oculta, letras_seleccionadas, intentos_restantes)`:
- Esta función imprime la información relevante del juego, como la palabra oculta, las letras seleccionadas y los intentos restantes.

### Función `seleccion_de_letra(letras_seleccionadas)`:
- Solicita al usuario que seleccione una letra.
- Valida si la entrada es una letra y si no ha sido seleccionada antes.
- Devuelve la letra seleccionada.

### Función `juego()`:
- Inicia el juego y establece las variables iniciales, como la palabra secreta, las letras acertadas, las letras seleccionadas, el número máximo de intentos y los intentos realizados.
- Ejecuta un bucle que continúa mientras no se hayan agotado los intentos.
  - Dentro del bucle, llama a `seleccion_de_letra()` para obtener la letra seleccionada.
  - Busca coincidencias de la letra seleccionada en la palabra secreta.
  - Imprime mensajes según si la letra está o no en la palabra secreta.
  - Actualiza la palabra oculta con las letras acertadas.
  - Llama a `imprimir_info()` para mostrar la información actualizada del juego.
  - Verifica si se ha adivinado la palabra completa o si se han agotado los intentos.

### Función `ocultar_palabra(palabra, letras_acertadas)` (repetida):
- Como se usa dentro de la función `juego()`, esta función se define nuevamente. Devuelve una cadena con letras ocultas y mostradas según las letras acertadas.

### Lógica del juego:
- El juego consiste en adivinar una palabra secreta que es un estado mexicano.
- El jugador tiene un número limitado de intentos (se establece en 6).
- En cada intento, el jugador selecciona una letra y se informa si está en la palabra secreta y en qué posiciones.
- La palabra secreta se muestra con asteriscos, revelando las letras adivinadas en las posiciones correctas.
- El juego continúa hasta que se adivine la palabra o se agoten los intentos.
